###ANALYSE SOCIO-DÉMOGRAPHIQUE

```python
# Sexe
sexe_counts = df['Sexe'].value_counts(normalize=True) * 100
print("\nRépartition par sexe (%):")
print(sexe_counts.to_string())

plt.figure()
sexe_counts.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Répartition par sexe')
plt.ylabel('Pourcentage')
plt.xticks(rotation=0)
plt.show()
```

```python
# Âge
print("\nStatistiques sur l'âge:")
age_stats = df['Age'].describe().to_string().split('\n')
for line in age_stats[:-2]:
    print(line)

plt.figure()
sns.histplot(df['Age'], bins=15, kde=True, color='teal')
plt.title('Distribution des âges')
plt.xlabel('Âge')
plt.show()
```

```python
# Commune
commune_counts = df['Commune'].value_counts(normalize=True) * 100
print("\nRépartition par commune (%):")
print(commune_counts.head(10).to_string())  # Affiche les 10 premières communes

plt.figure(figsize=(12, 6))
commune_counts.head(10).plot(kind='bar', color='lightgreen')
plt.title('Top 10 communes de résidence')
plt.ylabel('Pourcentage')
plt.xticks(rotation=45)
plt.show()
```

 ### ANALYSE COMPORTEMENTALE ET DE PERCEPTION

```python
# Connaissance des produits
connaissance_counts = df['Connaissance_Produits'].value_counts(normalize=True) * 100
print("\nConnaissance des produits existants (%):")
print(connaissance_counts.to_string())

```

```python
# Disponibilité perçue
Disponibilite_Produits_counts = df['Disponibilite_Produits'].value_counts(normalize=True) * 100
print("\nDisponibilité perçue (1-5):")
disponibilite_stats = Disponibilite_Produits_counts.describe().to_string().split('\n')
for line in disponibilite_stats[:-2]:
    print(line)
```

```python
   # Fréquence d'achat
freq_counts = df['Frequence_Achat'].value_counts(normalize=True) * 100
print("\nFréquence d'achat (%):")
print(freq_counts.to_string())

plt.figure()
freq_counts.plot(kind='bar', color='orange')
plt.title('Fréquence d\'achat')
plt.ylabel('Pourcentage')
plt.xticks(rotation=45)
plt.show()
```

```python
 # Dépense mensuelle
print("\nDépense mensuelle moyenne:")
depense_stats = df['Depense_Moyenne_Mois'].describe().to_string().split('\n')
for line in depense_stats[:-2]:
    print(line)

plt.figure()
sns.boxplot(y=df['Depense_Moyenne_Mois'], color='purple')
plt.title('Distribution des dépenses mensuelles')
plt.show()
```

```python
# Satisfaction globale
print("\nStatistiques sur la satisfaction globale (1-5):")
satisfaction_stats = df['Satisfaction_Globale'].describe().to_string().split('\n')
for line in satisfaction_stats[:-2]:
    print(line)

plt.figure()
sns.histplot(df['Satisfaction_Globale'], bins=5, discrete=True, color='royalblue')
plt.title('Distribution de la satisfaction globale')
plt.xlabel('Niveau de satisfaction (1-5)')
plt.show()
```

```python
 # Intérêt pour nouveau produit
interet_counts = df['Interet_Nouveau_Service'].value_counts(normalize=True) * 100
print("\nIntérêt pour un nouveau produit (%):")
print(interet_counts.to_string())
```

```python
# Supports d'information
support_counts = df['Support_Info'].value_counts(normalize=True) * 100
print("\nSupports d'information préférés (%):")
print(support_counts.to_string())

plt.figure()
support_counts.plot(kind='bar', color='lightcoral')
plt.title('Supports d\'information préférés')
plt.ylabel('Pourcentage')
plt.xticks(rotation=45)
plt.show()
```
